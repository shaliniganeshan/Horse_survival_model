```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
animals=pd.read_csv("horse.csv")
animals.head()
```

```python
target=animals['outcome']
target.head()
```

```python
target.unique()
```

```python
animals=animals.drop(['outcome'],axis=1) # out come column removed 
animals
```

```python
animals.info()
```

```python
catagory_variables=['surgery','age','temp_of_extremities',
                    'peripheral_pulse','mucous_membrane','capillary_refill_time',
                    'pain','peristalsis','abdominal_distention','nasogastric_tube',
                    'nasogastric_reflux','nasogastric_reflux_ph','rectal_exam_feces',
                    'abdomen','abdomo_appearance','surgical_lesion','cp_data'
                    ]
for catagory in catagory_variables:
    animals[catagory]=pd.get_dummies(animals[catagory])
    
animals[catagory].head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
X,y=animals.values,target.values
y = le.fit_transform(y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
print(X_train.shape)
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
imp=SimpleImputer(missing_values=np.nan,strategy="most_frequent")
```

```python
X_train=imp.fit_transform(X_train)
X_test=imp.fit_transform(X_test)
cls=DecisionTreeClassifier()
cls.fit(X_train,y_train)
```

```python
y_pre=cls.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_pre,y_test)
accuracy
```
